# Thought process for this section:
 - Check for which regions have nulls in the 'food_supply_(kcal/capita/day)' column
 - Create list of those regions
 - Create for loop to fill each regions nulls with the mean of the values for that region

## Food Supply

```python
df[df['food_supply_(kcal/capita/day)'].isnull()]
```

 ### Regions with nulls in food supply column
 1. Middle East and Northern Africa 
 2. Southern Asia
 3. Sub-Saharan Africa
 4. Central and Eastern Europe
 5. Western Europe
 6. Latin America and Caribbean
 7. Southeastern Asia


```python
regions = ['Middle East and Northern Africa', 'Southern Asia', 'Sub-Saharan Africa', 
            'Central and Eastern Europe', 'Western Europe', 'Latin America and Caribbean',
           'Southeastern Asia', 'Australia and New Zealand', 'Eastern Asia', 'North America']
```

```python
# df.loc[(df['region'] == 'Middle East and Northern Africa') & (df['food_supply_(kcal/capita/day)'].isnull()), 'food_supply_(kcal/capita/day)'] = df.loc[(df['region'] == 'Middle East and Northern Africa'), 'food_supply_(kcal/capita/day)'].mean()

# This was test code we used in the for loop below
```

```python
for i in regions:
    df.loc[(df['region'] == i) & (df['food_supply_(kcal/capita/day)'].isnull()), 'food_supply_(kcal/capita/day)'] = df.loc[(df['region'] == i), 'food_supply_(kcal/capita/day)'].mean()
```

```python
df.isnull().sum()
```

## Crude Birth Rate

```python
df[df['crude_birth_rate_(births_per_1,000_population)'].isnull()]
```

### Regions with nulls for crude birthrate
1. Sub-Saharan Africa 
2. Central and Eastern Europe
3. Western Europe


```python
for i in regions:
    df.loc[(df['region'] == i) & (df['crude_birth_rate_(births_per_1,000_population)'].isnull()), 'crude_birth_rate_(births_per_1,000_population)'] = df.loc[(df['region'] == i), 'crude_birth_rate_(births_per_1,000_population)'].mean()
    
```

```python
df.isnull().sum()
```
