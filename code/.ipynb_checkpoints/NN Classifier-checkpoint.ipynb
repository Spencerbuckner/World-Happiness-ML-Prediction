```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn import metrics

from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, Dropout, Input
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
```

```python
df = pd.read_csv('../Data/data_cleaned.csv')
df.drop(columns='Unnamed: 0', inplace=True)
df.head(2)
```

```python
country_to_numeric = {'Sub-Saharan Africa': 0, 'Central and Eastern Europe': 1, 'Latin America and Caribbean': 2, 
                     'Western Europe': 3, 'Middle East and Northern Africa': 4, 'Southeastern Asia': 5,
                     'Southern Asia': 6, 'Eastern Asia': 7, 'North America': 8, 'Australia and New Zealand': 9}

df['region'] = df['region'].replace(country_to_numeric)
```

```python
X = df.drop(columns=['year', 'region', 'country'])
y = df['region']
```

```python
X
```

```python
y
```

```python
y = to_categorical(y)
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
ss = StandardScaler()
ss.fit(X_train, y_train)
X_train_sc = ss.transform(X_train)
X_test_sc = ss.transform(X_test)
X_train.shape
```

```python
def model_func(layer_one_neurons, layer_two_neurons, dropout_percent):
    print('layer_one', layer_one_neurons, 'layer_two', layer_two_neurons) # only to help troubleshoot
    model = Sequential()
    model.add(Dense(layer_one_neurons, activation='relu', input_shape=(12,)))
    model.add(Dropout(dropout_percent))
    model.add(Dense(layer_two_neurons, activation='relu'))
    model.add(Dropout(dropout_percent))
    model.add(Dense(10, activation='softmax'))
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics='accuracy')
    return model

nn = KerasClassifier(build_fn=model_func, batch_size=8, verbose=0)

params = {
    'epochs': [50, 75, 100],
    'layer_one_neurons': [32, 64, 128],
    'layer_two_neurons': [32, 64, 128],
    'dropout_percent': [.25, .5]
}
gs = GridSearchCV(nn, param_grid=params, cv=5)
gs.fit(X_train_sc, y_train)
print(gs.best_score_) # <- negative mean squared error, ie the higher (less negative) the better
gs.best_params_
```

```python
model2 = Sequential()
model2.add(Dense(128, activation='relu', input_shape=(12,)))
model2.add(Dropout(.25))
model2.add(Dense(128, activation='relu'))
model2.add(Dropout(.25))
model2.add(Dense(10, activation='softmax'))
model2.compile(loss='categorical_crossentropy', optimizer='adam', metrics='accuracy')

results = model2.fit(X_train_sc, y_train,
         batch_size = 8,
         epochs=75,
         validation_data=(X_test_sc, y_test),
         verbose=0)
```

```python
preds = model2.predict(X_test_sc)
preds = preds.argmax(axis=1)
preds
```

```python
y_test_arg = y_test.argmax(axis=1)
y_test_arg
```

```python
from tensorflow.math import confusion_matrix
```

```python
cm = confusion_matrix(y_test_arg, preds)
plt.savefig('../visuals/nnm_confusion_matrix.png')
```
